```
a=['cast', 'homepage', 'director', 'tagline', 'keywords', 'overview']
df=df.drop(a,axis=1)
df.head()
```

```
b=df[df["release_year"].between(2000,2005)]
b["profit"]=(b["revenue"]-b["budget"])
b
```

```
#1)  Which movie had the most profit during 2000-2005?
profit=b["profit"].idxmax()
print("The movie that had the most profit during 2000-2005 is")
b.loc[[profit]]
```

```
df.sort_values("budget",ascending=True)
df
```

```
#2) Which are the movies with the second lowest and second highest budget?
re=0
min=df["budget"].min()
for index,value in enumerate(df["budget"],0):
    if value>min:
        re=index
        break
print("the second lowest budget movie is")
df.iloc[[re]]
```

```
re=0
min=df["budget"].max()
for index,value in enumerate(df["budget"],0):
    if value<min:
        re=index
        break
print("the second highest budget movie is")
df.iloc[[re]]
```

```
#3) Which are the movies with most and least earned revenue?
c=df["revenue"].idxmax()
print("The Least Revenue Earned Movie Is")
df.iloc[[c]]
```

```
c=df["revenue"].idxmin()
print("the most revenu earned movie is")
df.iloc[[c]]
```
